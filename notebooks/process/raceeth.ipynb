```python
import pandas as pd 
import numpy as np
```

```python
subtype = pd.read_csv('sustain_subtype_baseline_3x.csv')
demo = pd.read_csv('LEADS_PTDEMOG_08Jun2025.csv')
cn = pd.read_csv('full_gmm2z_exclude.csv')
cn = cn[cn.dx=='CN']
cn.drop_duplicates(subset=['subj'], inplace=True)
len(cn[cn.dx=='CN'].subj.unique())
```
