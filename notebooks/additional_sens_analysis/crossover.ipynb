```python
import pandas as pd 
import seaborn as sns 
from matplotlib import pyplot as plt
data_path = "/Users/mlin2/Desktop/RabLab/eoad_sustain_ml/data/"
```

```python
for i in pd.read_csv(data_path+"tableone_dat.csv").columns:
    print(i)
```

```python
df = pd.read_csv(data_path + "tableone_dat.csv")
print(len(df))
df = df[df.subtype_prob > 0.5]
sns.lineplot(
    data=df,
    x="stage",
    y="subtype_prob",
    hue="subtype",
    errorbar="sd"
)
plt.xlabel("Stage")
plt.ylabel("Subtype Probability")
plt.title("Subtype Probability by Stage")
plt.show()
```

```python
from matplotlib.colors import ListedColormap

# Define revised colormap
color_map = {
    1: "#1f449c",
    2: "#32ABA6",
    3: "#f05039",
}
 # Lineplot with error bars and custom color map, updated y-limit and x-ticks
sns.lineplot(
    data=df,
    x="stage",
    y="subtype_prob",
    hue="subtype",
    palette=color_map,
    errorbar="se"
)

# Add scatter plot in the background, colored by subtype
sns.stripplot(
    data=df,
    x="stage",
    y="subtype_prob",
    hue="subtype",
    palette=color_map,
    alpha=0.5,
    legend=False,
    jitter=0.2,
    size=4
)

plt.xlabel("Stage")
plt.ylabel("Subtype Probability")
plt.title("Subtype Probability by Stage")
plt.ylim(0.5, 1.05)
plt.xticks([0, 5, 10, 15, 20])
plt.show()
```
